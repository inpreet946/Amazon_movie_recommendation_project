Now let's grab the user ratings for those two movies:

```
Movie1_ratings = df['Movie127']
Movie101_user_ratings = df['Movie140']
Movie1_ratings.head()
```

```
Movie1_ratings.head()
```

```
similar_Movie1_ratings = df.corrwith(Movie1_ratings)
similar_Movie101_user_ratings= df.corrwith(Movie101_user_ratings)
```

```
corr_similar_Movie1_ratings = pd.DataFrame(similar_Movie1_ratings,columns=['Correlation'])
corr_similar_Movie1_ratings.dropna(inplace=True)
```

```
corr_similar_Movie1_ratings.head(5)
```

```
corr_similar_Movie101_user_ratings = pd.DataFrame(similar_Movie101_user_ratings,columns=['Correlation'])
corr_similar_Movie101_user_ratings.dropna(inplace=True)
```
